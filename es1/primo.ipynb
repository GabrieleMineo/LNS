Primi due esercizi: valore di <r> e varianza e loro incertezze montecarlo

```python
import numpy as np
import scipy as sp
import pandas as pd
import math
from matplotlib import pyplot as plt
def error(AV1,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV1[n] - AV2[n])/n)
    
M_ave, M2_ave, M_ave2 = np.loadtxt("out/first_prog.dat", usecols=(0, 1, 2), unpack='true')
V_ave, V2_ave, V_ave2 = np.loadtxt("out/second_prog.dat", usecols=(0, 1, 2),unpack='true')
N = len(M_ave)
n= np.arange(N)
e_m=np.zeros(N)
e_v=np.zeros(N)
for i in range(N):
    e_m[i] = error(M2_ave, M_ave2,i)
    e_v[i] = error(V2_ave, V_ave2,i)


plt.errorbar(n, M_ave-0.5, e_m);
plt.xlabel('#blocks')
plt.ylabel('$<x>-0.5$')
plt.grid(True)
plt.show()
plt.errorbar(n, V_ave-1/12, e_v);
plt.xlabel('#blocks')
plt.ylabel('$<(x-0.5)^2>-1/12$')
plt.grid(True)
plt.show()

```
