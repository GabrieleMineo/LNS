Nella seconda parte, sulle mosse gaussiane, vedremo come essenzialmente non si abbiano cambiamenti drastici nei risultati.

```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def error(a,b,n): 
    if n==0:
        return 0
    else:
        return math.sqrt((a[n] - b[n])/n)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y, z = np.loadtxt("out/1s.out", usecols= (0, 1, 2), delimiter = ' ', unpack = True)
d = len(x)
print('Orbitale (', x[0], y[0], z[0], ')')
ax.scatter(x, y, z, s= 0.1,c = z, marker = '.')
ax.view_init(1, 30)
plt.show()
fig= plt.figure()
r = np.loadtxt("out/r1s.out", usecols= (0, 1, 2), delimiter = ' ', unpack = True)
N = len (r[1,:])
x = np.arange(N)
std = np.zeros(N)
for i in range(N):
    std[i] = error(r[1,:], r[2,:], i)
   
plt.errorbar(x, r[0,:], std)
plt.xlabel('#blocks')
plt.ylabel('Raggio medio')
plt.title('Raggio medio (unitÃ  naturali)')
```
