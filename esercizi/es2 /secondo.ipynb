# Secondo Esercizio: Walker e suo andamento

di seguito riportiamo i risultti per l'andamento del nostro raggio medio del walker ripetendo $10^4$ volte un esperimento di 100 mosse su reticolo cubico

```python
from scipy import optimize
def test_func(x, parametro):
    return parametro*np.sqrt(x)

w_res = np.loadtxt("out/walker_file.dat")

x_=w_res[:,0]
y_=w_res[:,1]
z_=w_res[:,2]
params, params_covariance = optimize.curve_fit(test_func, x_, y_, p0=1.)

plt.errorbar(w_res[:,0],w_res[:,1],yerr=w_res[:,2])
plt.plot(x_,test_func(x_,params))

plt.show()
```

di seguito riportiamo invece il path per 1000 mosse del nostro walker

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x, y, z = np.loadtxt("out/walker_path.dat", usecols= (1, 2, 3), unpack = True)
d = len(x)
ax.plot3D(x, y, z, linewidth = 0.5, marker = ' ')
ax.scatter(0,0,0, marker='o', color ='g', label='start')
ax.scatter(x[d-1], y[d-1], z[d-1], marker = 'o', color = 'r', label = 'stop')
plt.legend()
plt.show()
```
