# Traveling Salesman Problem

abbiamo affrontato il problema direttamente con il Simulated annealing in parallelo, su quadrato e circonferenza per 70 citt√†.
abbiamo usato $1000$ temperature con scala lineare, per ogni temperatura abbiamo un certo numero di mosse provate $5000$ in particolare per i risultati presentati di seguito.
Mostriamo di seguito due risultati principali, uno che riguarda l'andamento temporale della lunghezza minima trovata, l'altro l'effettivo percorso compiuto.

```python
import random, numpy as np, math, copy, matplotlib.pyplot as plt

cities = np.loadtxt("out_c/posizioni.dat", unpack=True)
sequenze = np.loadtxt("out_c/conf_best.dat", unpack=True) #su ogni riga abbiamo la sequenza che ci interessa
tour = np.column_stack( (sequenze[:,1].astype(int),sequenze[:,-1].astype(int)) )
n_c=len(tour)

f = plt.figure(figsize = (12, 6))
for j in range(2):
    f.add_subplot(1, 2, j+1)
    plt.plot([cities[0][tour[i][j]] for i in range(n_c)], [cities[1][tour[i][j]] for i in range(n_c)], 'xb-');
plt.show()
```

```python
x, y, z = np.loadtxt("out_c/progressi.dat", unpack=True);

f = plt.figure(figsize = (12, 6))
for j in range(2):
    ax = f.add_subplot(1, 2, j+1)
    if(j==0): 
        plt.scatter(x, y);
        plt.title('length on circumference')
    else: 
        plt.scatter(x,z);
        plt.title('acceptance ratio circle')
    plt.xlabel('T')
    plt.xscale('log')
    plt.xlim(0.001, 10)
    
        
plt.show()
```
