### Exercise 10.1

By adapting your Genetic Algorithm code, developed during the Numerical Exercise 9, write a C++ code to solve the TSP with a **Simulated Annealing** (SA) algorithm. Apply your code to the optimization of a path among 

- 30 cities randomly placed on a circumference
- 30 cities randomly placed inside a square

Show your results via:

- a picture of the length of the best path as a function of the iteration of your algorithm
- a picture of the best path

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Cities on a circle

```python
read=open("RisultatiNotebook/30Cities/Circonferenza.dat")
NCity=read.readline()
NCity=read.readline()
```

```python
fig=plt.figure(figsize=(16,8))
fig.subplots_adjust(hspace=0.4, wspace=0.2)


plt.subplot(1,2,1)
x,y=np.loadtxt("RisultatiNotebook/30Cities/Risultati/circonferenza/posizioni.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Posizione delle città')
plt.scatter(x,y)
plt.grid(True)

plt.subplot(1,2,2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Primo percorso, random')
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
fig=plt.figure(figsize=(13,8))

beta,L=np.loadtxt("RisultatiNotebook/30Cities/Risultati/circonferenza/lunghezzeBest.dat", usecols=(0,1), delimiter='  ', unpack='true')
plt.plot(beta,L)
print("L_Best =", L[-1])
plt.xlabel('Beta')
plt.ylabel('L')
plt.title('Length as a function of 1/T')
plt.grid(True)

plt.show()
```

```python
read=open('RisultatiNotebook/30Cities/Risultati/circonferenza/sequenza.dat')
xBest=np.zeros(int(NCity)+1)
yBest=np.zeros(int(NCity)+1)
for i in range(int(NCity)):
    n=read.readline()
    xBest[i]=x[int(n)]
    yBest[i]=y[int(n)]
    
xBest[int(NCity)]=xBest[0]
yBest[int(NCity)]=yBest[0]
```

```python
fig=plt.figure(figsize=(10,8))
plt.plot(xBest,yBest)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Optimized Sequence')
plt.grid(True)

plt.show()
```
